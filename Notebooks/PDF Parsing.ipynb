```python
def filter_punct(text):
    text = text.lower()
    text = text.replace(',', '')
    text = text.replace(':', '')
    text = text.replace(';', '')
    text = text.replace('?', '')
    text = text.replace('<', '')
    text = text.replace('>', '')
    text = text.replace('.', '')
    text = text.replace('"', '')
    text = text.replace("'", '')
    text = text.replace('`', '')
    text = text.replace('\n', ' ')
    text = text.replace('-', ' ')
    text = text.split(' ')
    new_list = []
    for i in range(len(text)):
        if text[i] != '':
            new_list.append(text[i])
        else:
            pass
    return new_list
```

```python
txt = filter_punct(txt)
```

```python
def pdf_parser(filename):
    f = open(filename, 'rb')
    pdfObj = pdf.PdfFileReader(f)
    
    text = ''
    for page in range(pdfObj.numPages):
        new_page = pdfObj.getPage(page)
        new_text = new_page.extractText()
        text = text + new_text
    
    words_list = filter_punct(text)
    wc = Counter(words_list)
    num = len(words_list)
    f.close()
    return {'wordcount': wc, 'numwords': num}
```

```python
pdf_parser('sokal.pdf')
```
