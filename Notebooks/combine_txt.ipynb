```python
def combine_txt(txt_list):
    new_text = ''
    for txt in txt_list:
        print(txt)
        text_file = open(txt, encoding='utf-8')
        text = text_file.read()
        new_text = new_text + text

        text_file.close()
    return new_text
```

```python
drake_songs = ['Drake/cant_take_a_joke.txt', 'Drake/dont_matter_to_me.txt', 
            'Drake/nine.txt', 'Drake/nonstop.txt', 
            'Drake/redemption.txt', 'Drake/views.txt']
```

```python
drake_text = combine_txt(drake_songs)
```
