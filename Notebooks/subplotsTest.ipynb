```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
songs_df = pd.DataFrame()
songs_df['Artists'] = ['Drake', 'Drake', 'Kanye', 'Kanye', 'Kendrick', 'Kendrick']
songs_df['Album_name'] = ['Views', 'Scorpion', 'MBDTF', 'TCDO', 'MAAD', 'TPAB']
songs_df['Polarity'] = [0.009516460905349796, -0.02723531810070272, 0.02816805477131565, 0.018518497785739167, 0.023411003650134085, -0.04831967775581904]
songs_df['Subjectivity'] = [0.17124996804948492, 0.28417277167277166, 0.22448275456427627, 0.23512506012506032, 0.1723495280886586, 0.2212649528410397]
```

```python
print(songs_df)
```

```python
sns.barplot(data=songs_df.loc[songs_df['Artists'] == 'Drake'], x='Album_name', y='Polarity', hue='Album_name')
```

```python
sns.set()

fig, axes = plt.subplots(1, 3)

artists = ['Drake', 'Kendrick', 'Kanye']


for i in range(len(artists)):
    artist_name = artists[i]
    artist_albums = songs_df.loc[songs_df['Artists'] == artist_name]
        
    sns.barplot(data=artist_albums, x='Album_name', y='Subjectivity', hue='Album_name', ax=axes[i])
    axes[i].set_title(artist_name)
    plt.subplots_adjust(wspace=0.75)
    plt.gcf().set_size_inches((10, 5))
    axes[i].set_ylim(0, 0.3)
    
```
